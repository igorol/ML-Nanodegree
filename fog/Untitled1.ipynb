```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dateparse = lambda x: pd.datetime.strptime(x, '%Y-%m-%d %H:%M:%S')
data_saez = pd.read_csv('SAEZ.csv',low_memory=False, parse_dates=['valid'], date_parser=dateparse)
data_saez = data_saez.dropna(0,how='any')

data_sumu = pd.read_csv('SUMU.csv',low_memory=False, parse_dates=['valid'], date_parser=dateparse)
data_sumu = data_sumu.dropna(0,how='any')
```

```python
plt.rcParams['figure.figsize'] = (16,10)
plt.rcParams.update({'font.size': 18})
plt.scatter(data_saez['valid'].values,data_saez['vsby'], color='blue')
plt.axhline(y=1, color='r', linestyle='-')
plt.ylabel('Visibility (Km)')
plt.title('SAEZ (Buenos Aires) Airport - Visibility Reports (Km)')
plt.savefig('vsby_saez.png')
plt.show()
```

```python
plt.scatter(data_sumu['valid'].values,data_sumu['vsby'], color='blue')
plt.axhline(y=1, color='r', linestyle='-')
plt.ylabel('Visibility (Km)')
plt.title('SUMU Airport (Montevideo) - Visibility Reports (Km)')
plt.savefig('vsby_sumu.png')
plt.show()
```
