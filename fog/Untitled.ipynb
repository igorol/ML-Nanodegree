```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def analyze(airport):
    dateparse = lambda x: pd.datetime.strptime(x, '%Y-%m-%d %H:%M:%S')
    data = pd.read_csv('{}.csv'.format(airport),low_memory=False, parse_dates=['valid'], date_parser=dateparse)
    data = data.dropna(0,how='any')

    plt.clf()
    plt.figure(figsize=(15,5))
    plt.plot(data.valid,data.vsby,'bo',markersize=1)
    plt.axhline(y=1, color='r', linestyle='-')
    plt.title('{} Airport'.format(airport))
    plt.savefig('vsby_{}.png'.format(airport))
    plt.show()
    
    
    print np.histogram(data['vsby'], [0,1.01,7])
    
    
    
```

```python
analyze('SUMU')
```
